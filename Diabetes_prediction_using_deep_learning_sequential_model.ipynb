```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/bansalrishi/MachineLearningWithPython_UD/master/Data/Deep_Learning/pima-data.csv', delimiter=',')
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['diabetes'] = le.fit_transform(data['diabetes'])
```

```python
data.head()
```

```python
x=data[['num_preg','glucose_conc','diastolic_bp','insulin','bmi','diab_pred','age','skin']]
y=data['diabetes']
```

```python
#creating the model
model = Sequential()

#adding first dense layer with 8 independent feature or dimensions, taking relu as activation function
model.add(Dense(12, input_dim=8, activation='relu'))

#adding second dense layer with 8 nodes to recieve outputs from first dense layer, taking relu as activation function
model.add(Dense(8, activation='relu'))

#adding third dense layer with 1 nodes to recieve output fro second dense layer , taking sigmoid as activation function to classify the result between 0 and 1
model.add(Dense(1, activation='sigmoid'))
```

```python
#compiling the model
#using binary_crossentropy as loss function
#using adam optimizer to optimize
#metrics are based on accuracy of classification
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
data.info()
```

```python
#as mentioned below 150 epocs or iterations 
#and batch size is 10
model.fit(x, y, epochs=150, batch_size=10)
```

```python
accuracy = model.evaluate(x, y)
print('Accuracy: ', (accuracy[1]*100),'%')
```
